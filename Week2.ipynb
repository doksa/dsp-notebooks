```python
%matplotlib inline

import numpy as np
import scipy.io.wavfile as wav
from scipy.signal import resample
from IPython.display import Audio
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
# Approximating a square wave by partial sums of its Fourier series

T = 1
om0 = 2*np.pi / T
t = np.arange(-T, 3*T, .0005)

# number of odd harmonics to add together
Q = 5 

# compute the approximation
xx = np.zeros(t.shape)
for l in range(Q):
    xx += 4/np.pi * 1 / (2*l + 1) * np.sin((2*l + 1)*om0*t)


# do some plotting    
fig = plt.figure(figsize=(7, 4))

plt.plot(t, np.sign(np.sin(om0*t))) # square wave
plt.plot(t, xx)                     # approximation

plt.xlim(-.1, 2.1)
plt.ylim(-1.5, 1.5)




```
