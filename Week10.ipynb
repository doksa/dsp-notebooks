```python
# Phase responses of various filters

%pylab inline

fig = plt.figure(figsize=(12, 5))

N = 1024
w = np.linspace(-np.pi, np.pi, N)

ax = fig.add_subplot(131)
h = np.array([1, 0, 2, 3, 2, 0, 1])
plt.plot(w, np.fft.fftshift(np.angle(np.fft.fft(h, N))))

ax = fig.add_subplot(131)
h = np.array([1, 0, 2, 0, -2, 0, -1])
plt.plot(w, np.fft.fftshift(np.angle(np.fft.fft(h, N))))

ax = fig.add_subplot(131)
h = np.array([1, 0, 2, 3, -2, 0, -1])
plt.plot(w, np.fft.fftshift(np.angle(np.fft.fft(h, N))))
```
